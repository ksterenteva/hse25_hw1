## Находим пересечение с генами по координатам

```python
!bedtools intersect -a DMR_results_8cell_icm_chr11.bed -b RefSeqGenes_chr11_fixed.bed -wa > DMR_in_genes_8cell_icm_chr11.bed
```

```python
import pandas as pd

# Загружаем файл с аннотацией генов
genes_df = pd.read_csv("RefSeqGenes_mm39.bed", sep="\t", header=None)

# Проверяем, что загружено
print("Структура файла:")
print(genes_df.head())

# Фильтруем только 11-ю хромосому
genes_chr11 = genes_df[genes_df.iloc[:, 1] == "chr11"]

# Убеждаемся, что координаты целые
genes_chr11.iloc[:, 2] = genes_chr11.iloc[:, 2].astype(int)
genes_chr11.iloc[:, 3] = genes_chr11.iloc[:, 3].astype(int)

# Сохраняем в правильном BED-формате (без заголовков!)
genes_chr11.to_csv("RefSeqGenes_chr11_fixed.bed", sep="\t", index=False, header=False)

print("✅ Файл RefSeqGenes_chr11_fixed.bed сохранен в BED-формате!")
```

```python
import pandas as pd

dmr_file = "DMR_in_genes_chr11.bed"
genes_file = "RefSeqGenes_chr11_final.bed"

dmr_df = pd.read_csv(dmr_file, sep="\t", header=None, names=["chr", "start", "end"])
genes_df = pd.read_csv(genes_file, sep="\t", header=None, names=["gene", "chr", "gene_start", "gene_end"])

dmr_df = dmr_df[dmr_df["chr"] == "chr11"]
genes_df = genes_df[genes_df["chr"] == "chr11"]

def find_gene(dmr_start, dmr_end):
    match = genes_df[(genes_df["gene_start"] <= dmr_end) & (genes_df["gene_end"] >= dmr_start)]
    if not match.empty:
        return ",".join(match["gene"].tolist())
    return "No_Gene"

dmr_df["gene"] = dmr_df.apply(lambda row: find_gene(row["start"], row["end"]), axis=1)

output_file = "Final_DMR_in_Genes_chr11.bed"
dmr_df.to_csv(output_file, sep="\t", index=False, header=False)
```
